```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt
```

# NLTK Gutenberg Corpus

```python
texts = nltk.corpus.gutenberg.fileids()
print(f"There are {len(texts)} books in the NLTK Gutenberg Corpus")
```

```python
# create a dicitonary of books and their words grouped by sentence
sentence_dict = {}
for book in texts:
    sentence_dict[book] = nltk.corpus.gutenberg.sents(book)

```

## How many sentences are there per book?

```python
num_sent_per_book = [len(sentence_dict[book]) for book in sentence_dict]
```

```python
fig, ax = plt.subplots()
ax.barh( list(sentence_dict.keys()), num_sent_per_book,)
```

The Bible has a lot of sentences compared to the other texts.  
Blake's poems have very few. 

The wide range of sentence count could present a class imbalance problem for a machine learning model that categorizes new sentences.  If the imbalance were left as is, the model may end up always predicting a sentence is biblical.
